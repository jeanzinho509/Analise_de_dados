```python
#Aqui criamos um novo Dataframe apenas com as colunas que eu quero
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
#Exibi a 4a. linha
display(ca_df.loc[3])
```

```python
#exibi a 10. linha até 20a. linha
display(ca_df.loc[9:19])
```

```python
#Criar uma dataframe gas_df 
#contendo apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
#Somente com combustivel sendo GASOLINA e exibir na tela
gas_df = ca_df.loc[ca_df['Produto']== 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o posto e o municipio da gasolina mais cara do pais?
display(gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com multiplas condiçoes para filtragem
 #Quais são os preços, postos que vendem ETANOL na minha cidade (RIO DE JANEIRO)?
 #ordenado do menor preço para o maior
etanol_rj_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'RIO DE JANEIRO')]
display(etanol_rj_df.sort_values(by='Valor de Venda'))
```

```python
#qual a media de preços dos combustiveis GASOLINA e GASOLINA ADITIVADA do Baireo BELA VISTA em SÃO PAULO
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), ['Valor de Venda']].mean())
```

```python
#qual a média de preços dos combustiveis GASOLINA e GASOLINA ADITIVADA do Baireo MOOCA em SÃO PAULO
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') &
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin([ "GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL Brasil todo?
media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```
