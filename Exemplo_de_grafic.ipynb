```python
#Criando um pequeno gráfico

import pandas as pd

import matplotlib.pyplot as plt

num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
display(num_habitantes_df)
print(num_habitantes_df.info())
```

```python
#verificação visual rápida dos dados de NumHabitantes2021.

num_habitantes_df.plot()

plt.show()
```

```python
#Verificando os Estados mais Habitados e Menos Habitados
import numpy as np

num_habitantes_df['Status do Estado'] = np.where(num_habitantes_df['NumHabitantes2021'] > 2000000, 'Mais Habitado', 'Menos Habitado')
display(num_habitantes_df[['Municipio', 'Estado', 'NumHabitantes2021', 'Status do Estado']])
```

```python
#Criando um gráfico mostrando os Estados mais Habitados

# Grupando data por estado e calcular a soma dos habitantes
state_group = num_habitantes_df.groupby('Estado')['NumHabitantes2021'].sum()

# Traçar a soma dos habitantes por estado
state_group.plot(kind='bar', figsize=(10,6), color='blue')

# Adicione rótulos e título ao enredo
plt.xlabel('Estado')
plt.ylabel('Numero de Habitantes')
plt.title('Numero de Habitantes por Estado')

# Mostrar
plt.show()
```
