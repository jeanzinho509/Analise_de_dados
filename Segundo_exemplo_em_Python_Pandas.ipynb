```python
#Importando biblioteca pandas no Python
import pandas as pd
```

```python
combustiveis_df = pd.read_excel("ca202102.xlsx")
display(combustiveis_df.head())
```

```python
#Inserção simples de dado
combustiveis_df['Ativo'] = True

display(combustiveis_df.head())
```

```python
from numpy.lib.function_base import disp
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE" quando a coluna Municipio for igual a RIO DE JANEIRO
combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'RIO DE JANEIRO' else '' for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['RIO DE JANEIRO','NITEROI', 'ANGRA DOS REIS', 'SAQUAREMA']), ['Municipio', 'Obs']])
```

```python
#como preencher uma coluna 'Valor de Venda - Status' que verifica o seguinte :
# Se o valor de venda for maior que 6,5 reais, ele fala que ta Caro.. caso contrario, está barato
import numpy as np

combustiveis_df['Status do Valor de Venda'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Status do Valor de Venda']])
```

```python
#Calcular postos por habitante temos na mostragem de combustiveis nov/2021

habitantes_city_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
habitantes_city_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(habitantes_city_df)
```

```python
#Fazer um MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(habitantes_city_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Destruir coluna completamente vazia (todas as linhas são nulas)
merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Status do Valor de Venda']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
#Remoer as linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#Agrupar e contar quantos postos tem na cidade.
postos_por_munip_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_munip_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_munip_df.rename(columns={"Revenda": "Número de Postos"}, inplace=True)
display(postos_por_munip_df)
```

```python
#Agrupar e contar quantos postos tem na cidade
postos_por_munip_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_munip_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_munip_df.rename(columns={'Revenda': 'Numero de Postos'}, inplace=True)
postos_por_munip_df.reset_index(inplace=True)
display(postos_por_munip_df)
```

```python
#Agrupar e contar quantos postos tem na cidade..
postos_por_munip_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_munip_df.reset_index(inplace=True)
#display(postos_por_municipio_df.info())
postos_por_munip_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_munip_df.rename(columns={"Revenda": "NumPostos"}, inplace=True)

postos_por_munip_df['NumHabitantesPorPosto'] = postos_por_munip_df['NumHabitantes2021'] / postos_por_munip_df['NumPostos']
#display(postos_por_municipio_df.info())
display(postos_por_munip_df)
```

```python
#Visualização do consumo médio
c_mean = combustiveis_df[ 'Valor de Venda' ].groupby(by=combustiveis_df['Produto']) .mean ()
display(c_mean)
```

```python
#Vou definir a área do gráfico
plt.figure(figsize=(7,5))

#Plotar o gráfico
c_mean.plot(
    kind="barh",
    xlabel="Tipo de Combustível",
    ylabel="Preço reais/litro",
    color="red"
)

#Grid
plt.grid()

#Exibe
plt.show()
```

```python
import seaborn as sns

#Vou definir a área do gráfico
plt.figure(figsize=(7,5))

#Plotar o gráfico
c_mean.plot(
    kind="barh",
    xlabel="Tipo de Combustível",
    ylabel="Preço reais/litro",
    title="Média de preços por combustível",
    color="red",
    alpha=0.3
)

#Grid
plt.grid()

#Remover as linhas superior e lateral direita do gráfico 
sns.despine()

#Exibe
plt.show()
```
